```python
# Buka filenya
file = open('doc.txt', encoding='utf8')
read = file.read()
file.seek(0)
read

# Menghitung Jumlah baris
line = 1
for word in read:
	if word == '\n':
		line += 1
print("Jumlah Baris dalam Dokumen tersebut adalah: ", line)

#Membuat list untuk menyimpan tiap baris sebagai elemen daripada list
array = []
for i in range(line):
	array.append(file.readline())

array
```

```python
# Menghapus tanda yang tidak perlu
punc = '''!()-[]{};:'"\, <>./?@#$%^&*_~'''
for ele in read:
	if ele in punc:
		read = read.replace(ele, " ")
		
read

# Membuat Tulisan menjadi lowercase
read=read.lower()					
read
```

```python
# Mengimpor module nltk dan kawan-kawan
from nltk.tokenize import word_tokenize
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
nltk.download('punkt')
```

```python
for i in range(1):
	# Konversi menjadi token
	text_tokens = word_tokenize(read)

tokens_without_sw = [
	word for word in text_tokens if not word in stopwords.words()]

print(tokens_without_sw)
```

```python
# Dictionary Inverted Index
dict = {}

for i in range(line):
	check = array[i].lower()
	for item in tokens_without_sw:

		if item in check:
			if item not in dict:
				dict[item] = []

			if item in dict:
				dict[item].append(i+1)

dict
```
